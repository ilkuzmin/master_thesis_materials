```python
# Задаем название будущим колонкам

cols = [
    'Регион отбытия',
    'Всего мигрантов в возрасте 14 лет и более',
    'Военно-служащие',
    'Руководители',
     'Специалисты высшего уровня квалификации, всего',
     'Специалисты ВУК в области науки и техники',
     'Специалисты ВУК в области здраво-охранения',
     'Специалисты ВУК в области образования',
     'Специалисты ВУК в сфере бизнеса и администрирования',
     'Специалисты ВУК по инфор-мационно-коммуникационным технологиям (ИКТ)',
     'Специалисты ВУК в области права, гуманитарных областей и культуры',
     'Специалисты среднего уровня квалификации всего',
     'В области науки и техники',
     'Средний медицинский персонал здравоохранения',
     'Специалисты (ср.квал) по экономической и административной деятельности',
     'Специалитсы (ср. квал) в области правовой, социальной работы, культуры, спорта и родственных занятий',
     'Специалитсы (ср. квал) в области информационно-коммуникационным технологиям (ИКТ)',
     'Служащие, занятые подготовкой и оформлением документации, учетом и обслуживанием, всего',
     'Служащие общего профиля и обслуживающие офисную технику',
     'Служащие сферы обслуживания населения',
     'Служащие в сфере обработки числовой информации и учета материальных ценностей',
     'Работники сферы обслуживания и торговли, охраны граждан и собственности',
     'Квалифицированные работники сельского и лесного хозяйства, рыбоводства и рыболовства',
     'Квалифицированные рабочие промышленности, строительства, транспорта и рабочие родственных занятий',
     'Операторы производственных установок и машин, сборщики и водители',
     'Неквалифицированные рабочие',
     'Не указано'
    ]
```

```python
# Форма, куда все сводим
data_empty = []
data_migr_z = pd.DataFrame(data = data_empty, columns = cols)

data_migr_z.insert(1,'Регион прибытия','') #по какому региону табличку...
data_migr_z.insert(2,'Год','') #по какому региону табличку...

data_migr_z.head()
```

```python
# основной скрипт

for migr in ['Мигр2019', 'Мигр2020']:
    for dir in os.listdir(os.getcwd() + '\{}'.format(migr))[:8]:
        print('Working witn {}, {}'.format(migr, dir))
        
        workbook = xlrd.open_workbook('D:/Архивы Россттата/{}/{}/mz2_reg.xls'.format(migr, dir))

        sheets_list = [sheet.name for sheet in workbook.sheets()]
        iter_list = sheets_list[6::6]
        
        for i in iter_list:

            worksheet = workbook.sheet_by_name(i)
            subj = worksheet.cell(1,0).value

            year = int(worksheet.cell(0,0).value[-9:-5])

            data = pd.read_excel('D:/Архивы Россттата/{}/{}/mz2_reg.xls'.format(migr, dir),
                                sheet_name=i, index_col = 0, skiprows = 6 
                                )
            data.columns = cols

            data.insert(1,'Регион прибытия','') #по какому региону табличку...
            data['Регион прибытия'] = [subj]*len(data) 

            data.insert(2,'Год','') #...и в каком году
            data['Год'] = [year]*len(data)
            data_migr_z = pd.concat([data_migr_z,data], ignore_index = True)
```
