## Сначала сводим данные, которые сгруппированы по папкам. Получим 3 года наблюдений (2016, 2018, 2019) 

```python
# скрипт для 2019 года

for migr in ['Мигр2019']:
    for dir in os.listdir(os.getcwd() + '\{}'.format(migr))[:8]:
        print('Working witn {}, {}'.format(migr, dir))
        
        workbook = xlrd.open_workbook('D:/Архивы Россттата/{}/{}/mo3_reg.xls'.format(migr, dir))

        sheets_list = [sheet.name for sheet in workbook.sheets()]
        iter_list = sheets_list[9::9]
        
        for i in iter_list:

            worksheet = workbook.sheet_by_name(i)
            subj = worksheet.cell(1,0).value

            year = [int(s) for s in worksheet.cell(0,0).value.split() if s.isdigit()][-1]

            data = pd.read_excel('D:/Архивы Россттата/{}/{}/mo3_reg.xls'.format(migr, dir),
                                sheet_name=i, index_col = 0, skiprows = 6 
                                )
            data.columns = cols

            data.insert(1,'Регион прибытия','') #по какому региону табличку...
            data['Регион прибытия'] = [subj]*len(data) 

            data.insert(2,'Год','') #...и в каком году
            data['Год'] = [year]*len(data)
            data_migr_edu = pd.concat([data_migr_edu,data], ignore_index = True)
```

```python
# скрипт по 2016 и 2018 годам 
for migr in ['Мигр2016', 'Мигр2018']:
    for dir in os.listdir(os.getcwd() + '\{}'.format(migr))[:8]:
        print('Working witn {}, {}'.format(migr, dir))
        
        workbook = xlrd.open_workbook('D:/Архивы Россттата/{}/{}/mo3_reg.xls'.format(migr, dir))

        sheets_list = [sheet.name for sheet in workbook.sheets()]
        iter_list = sheets_list[15::9] # ИЗ: тут по сравнению с 19 годом другой порядок 
        
        for i in iter_list:

            worksheet = workbook.sheet_by_name(i)
            subj = worksheet.cell(1,0).value

            year = [int(s) for s in worksheet.cell(0,0).value.split() if s.isdigit()][-1]

            data = pd.read_excel('D:/Архивы Россттата/{}/{}/mo3_reg.xls'.format(migr, dir),
                                sheet_name=i, index_col = 0, skiprows = 6 
                                )
            data.columns = cols
            data = data[:97] # эта строчка убирает данные по международной миграции, которая нам не нужна

            data.insert(1,'Регион прибытия','') #по какому региону табличку...
            data['Регион прибытия'] = [subj]*len(data) 

            data.insert(2,'Год','') #...и в каком году
            data['Год'] = [year]*len(data)
            data_migr_edu = pd.concat([data_migr_edu,data], ignore_index = True)
```

```python
# 2015 год 
for migr in ['Мигр2015']:
    list_files = os.listdir('D:/Архивы Россттата/{}'.format(migr))[:-1]
    for file in list_files:
        print('Working witn {}, {}'.format(migr, file))
        
        workbook = xlrd.open_workbook('D:/Архивы Россттата/{}/{}'.format(migr, file))
        
        sheets_list = [sheet.name for sheet in workbook.sheets()]  
        r = re.compile(r'(?i)mo3_\d+_3_1')
        i = list(filter(r.match, sheets_list))[0]


        worksheet = workbook.sheet_by_name(i)
        subj = worksheet.cell(1,0).value

        year = [int(s) for s in worksheet.cell(0,0).value.split() if s.isdigit()][-1]

        data = pd.read_excel('D:/Архивы Россттата/{}/{}'.format(migr, file),
                            sheet_name=i, index_col = 0, skiprows = 6, na_values = '-'
                            )
        data.columns = cols
        data = data[:98].fillna(0) # эта строчка убирает данные по международной миграции, которая нам не нужна, и заполняет пропуски

        data.insert(1,'Регион прибытия','') #по какому региону табличку...
        data['Регион прибытия'] = [subj]*len(data) 

        data.insert(2,'Год','') #...и в каком году
        data['Год'] = [year]*len(data)
        data_migr_edu = pd.concat([data_migr_edu,data], ignore_index = True)
```
